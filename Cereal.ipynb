Ahora procedemos a obtener el producto con mas calorias de cada marca de producto.

```python
#Data type of the Dataset
df.dtypes
```

```python
df.groupby(['mfr'], sort=False)['calories'].max()
#Now we obtain the product that correspond the max calorie
idx = df.groupby(['mfr'])['calories'].transform(max) == df['calories']
df[idx]
```

We can see that exist some brands with more than one product with same level of calories
